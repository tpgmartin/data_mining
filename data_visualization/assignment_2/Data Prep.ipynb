```python
import numpy as np
import pandas as pd

facebook = pd.read_csv("facebook.csv")

print(facebook)

data = {}
```

```python
all_nodes = np.union1d(facebook["nodes"],facebook["links"])
```

```python
data["nodes"] = []

for node in all_nodes:
    if node < 100:
        data["nodes"].append({"id": node})

data["nodes"]
```

```python
data["links"] = []

for i in range(facebook.shape[0]):
    if facebook.iloc[i].nodes < 100 and facebook.iloc[i].links < 100:
        data["links"].append({"source": facebook.iloc[i].nodes,"target": facebook.iloc[i].links,"weight":1})

data["links"]
```

```python
%store data > facebook.json
```
