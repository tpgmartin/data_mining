```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('ExcelFormattedGISTEMPData2CSV.csv')
```

```python
data.head()
```

```python
data.tail()
```

```python
plt.figure(figsize=(10,6))
plt.plot(data.iloc[:,0], data.iloc[:,2:3], label="Northern Hemisphere")
plt.plot(data.iloc[:,0], data.iloc[:,3:4], label="Southern Hemisphere")
plt.title('Temperature Deviations Northern Hemisphere vs Southern Hemisphere 1880-2014', size=20)
plt.xlabel('Year', size=14)
plt.ylabel('Deviation (0.01 Degree Celsius)', size=14)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

The plot above demonstrates the change in temperature deviations over the period 1880-2014, with the years running along the x-axis and the temperature deviation along the y-axis. The chart uses the historic data for the Northern and Southern hemispheres only.

There a some insteresting features and trends to this chart

* A general and significant upward trend in the average temperature deviation over the time period shown, for both hemispheres
* Temperature deviations increased more noticeably and consistently from around 1970
* Temperature deviations increased more noticeably in the northern hemisphere

Overall this chart shows a overall increase in temperature deviations over time across both hemispheres. The upward trend indicates a rise in global temperatures.
